```python
sns.set(style='darkgrid', palette='twilight_shifted', color_codes=True, rc={'figure.figsize':(15,10)})

filtered_Genre = ['action', 'comedy', 'adventure', 'animation', 'drama']
movies_filt1 = movies[movies.Genre.isin(filtered_Genre)]

filtered_Studio = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
movies_filt2 = movies_filt1[movies_filt1.Studio.isin(filtered_Studio)]

movies_filtered = movies_filt2
len(movies_filtered)
```

```python
g = sns.boxplot(data=movies_filtered, x='Genre', y='Gross%US', orient='v', color='lightgray', showfliers=False)
plt.setp(g.artists, alpha=0.5)

sns.stripplot(data=movies_filtered, x='Genre', y='Gross%US', jitter=True, size=6, hue='Studio',linewidth=0, \
              alpha=0.7)

g.axes.set_title('Domestic Gross % by Genre', fontsize =30)
g.set_xlabel('Genre', fontsize=20)
g.set_ylabel('Gross % US', fontsize=20)
g.set(ylim= (0,100))
g.legend(bbox_to_anchor=(1.05,1), loc=2)
```
